```python
# load the csv
import pandas as pd

df= pd.read_csv('Meteorite_Landings.csv')
df.head(5)
```

```python
# change the date of the year to only show the year

df['year'] = df.year.str.slice(6, 10)
df

```

```python
# convert year to a numeric data type

df.year = df.year.apply(pd.to_numeric)
df
```

```python
# fill missing values with zero

df = df.fillna(0)
```

```python
df.isna().sum()
```

```python
# new column whether meteorite was observed falling before 1970

df['observed'] = (df['year'] < 1970)
df
```

```python
# set the index to to the id column and extrac all rows between 10,036 and 10,041
df = df.set_index('id')
df = df.sort_index()
df
```
