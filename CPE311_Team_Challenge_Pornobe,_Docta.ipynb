<h2>
*   Let's Pretend we're on the metric system and use kilogram instead of pounds to give us a starting base number of 100.<br>
*   Four people (Roman, Verlyn, Lloyd, and Robin) want to cross a river in a boat that can only carry 100kg. <br>
*   Roman weighs 90kg, Verlyn weighs 80kg, Lloyd weighs 60kg and Robin weighs 40kg, and they have 20kg of supplies.<br>
*   How do they get cross?





Final Code




```python
class Passenger:
    def __init__(self, name, weight, living,place):
        self.name = name
        self.weight = weight
        self.living = living
        self.place = place


    def move(self, newPlace):
        self.place = newPlace

    def show(self):
        print(self.name, self.weight, "kg")

class river_crossing():
    def __init__(self):
        self.p1 = Passenger("Roman", 90, True, "Origin")
        self.p2 = Passenger("Verlyn", 80, True, "Origin")
        self.p3 = Passenger("Lloyd", 60, True, "Origin")
        self.p4 = Passenger("Robin", 40, True,  "Origin")
        self.p5 = Passenger("Supply", 20, False,  "Origin")

    class solution():
        def __init__(self):
            self.passengers = [river_crossing().p1, river_crossing().p2, river_crossing().p3, river_crossing().p4, river_crossing().p5]
            self.originList = []
            self.boatList = []
            self.destinationList = []
            self.seperator = []


        # ==========================================================================================================
        # check if the specified place is empty
        def checkEmpty(self, place):
            temp = 0
            for i in self.passengers:
                if i.place == place:
                    temp += 1
            if temp > 0:
                return False
            else:
                return True

        # for showing all the current passengers at each places
        def showPlace(self):
            for i in self.passengers:
                if i.place == "Origin":
                    self.originList.append(i)

                elif i.place == "Boat":
                    self.boatList.append(i)

                elif i.place == "Destination":
                    self.destinationList.append(i)

            print("Origin: ", end =' ')
            for i in self.originList:
                print(i.name, i.weight, "kg", end = ', ')
            print()

            print("Boat: ", end = '')
            for i in self.boatList:
                print(i.name, i.weight, "kg", end = ', ')
            print()

            print("Destination: ", end = '')
            for i in self.destinationList:
                print(i.name, i.weight, "kg", end = ', ')
            print()
            print("=============================================================================")

            # clean the list for the next update
            self.originList.clear()
            self.boatList.clear()
            self.destinationList.clear()

        # to check the if the boat is full
        def checkBoat(self, originalDestination):
            temp = 0
            check = 0

            # compute the total weight
            for i in self.passengers:

                if i.place == "Boat":
                    temp += i.weight

            # if boat is full dont accept any passenger
            if temp == 100:
                return False

            elif temp <= 100:
                # check if it is possible to add another passenger to the boat from the origin
                for i in self.passengers:
                    if i.place == originalDestination:
                        temp += i.weight
                        if temp <= 100:
                            check += 1

                if check > 0:
                    return True

        # to move a passenger to the boat
        def travelToBoat(self, originalDestination, targetDestination):
            max_weight_exist = False
            print("Went to boat: ")
            # check first if there is a combination that exactly at the boat limit
            if max_weight_exist == False:
                for i in self.passengers:
                    for j in self.passengers:
                        if i != j and self.checkBoat("Origin") and i.place == originalDestination and j.place == originalDestination:
                            if i.weight + j.weight == 100:
                                i.move(targetDestination)
                                j.move(targetDestination)
                                i.show()
                                j.show()
                                print("=============================================================================")
                                max_weight_exist = True

            # if there is none proceed to other combination which is less than boat limit
            elif max_weight_exist == False:
                for i in self.passengers:
                    for j in self.passengers:
                        if i.weight + j.weight < 100:
                            i.move(targetDestination)
                            j.move(targetDestination)
                            i.show()
                            j.show()
                            print("=============================================================================")

            if self.checkBoat("Origin"):
                heaviest = 0
                for i in self.passengers:
                    if i.place == originalDestination:
                        if heaviest < i.weight:
                            heaviest = i.weight

                for i in self.passengers:
                    if i.weight == heaviest:
                        i.move("Boat")
                        i.show()
                        print("=============================================================================")

        def travelBackToBoat(self):
            print("Went back to boat: ")
            lightest = None
            if self.checkBoat("Destination"): # check if it is possible to add a passenger into the boat from the Destination
                for i in self.passengers:
                    if i.living and i.place == "Destination":
                        if lightest == None or i.weight < lightest: # if there is no lightest yet, let lightest be the first weight of passenger or if there is already one, compare
                            lightest = i.weight

            for i in self.passengers: # find the lightest in the list (could be improved or optimized this part)
                if i.weight == lightest:
                    i.move("Boat")
                    i.show()
                    print("=============================================================================")

        # move the heaviest passenger to the destination
        def travelToDestination(self):
            for i in self.passengers:
                if i.place == "Boat":  # if the place of an object is the boat, put in a list
                    self.seperator.append(i)

            print("Went to Destination: ")
            for i in self.seperator:
                for j in self.seperator:

                    # if supply automatically go to destination
                    if i.living == False:
                        i.move("Destination")
                        i.show()
                        print("=============================================================================")
                        break

                    # if supply automatically go to destination
                    elif j.living == False:
                        j.move("Destination")
                        j.show()
                        print("=============================================================================")
                        break
                    # if i is heavier move to destination
                    elif i.weight > j.weight:
                        i.move("Destination")
                        i.show()
                        print("=============================================================================")
                        break
                    # if j is heavier move to destination
                    elif i.weight < j.weight:
                        j.move("Destination")
                        j.show()
                        print("=============================================================================")
                        break
                    # if only one element is in the boat move to destination
                    elif len(self.seperator) == 1:
                        i.move("Destination")
                        i.show()
                        print("=============================================================================")
                break
            self.seperator.clear()


        def travelToOrigin(self):
            print("went back to Origin: ")
            for i in self.passengers:
                if i.place == "Boat":
                    i.move("Origin")
                    i.show()
                    print("=============================================================================")


    def execute(self):
            k = 1
            solution = self.solution()
            while solution.checkEmpty("Origin") == False: # while there is still a passenger at origin, execute the algo
                if solution.checkBoat("Origin"): # if boat is not full put passengers in the boat


                    print("Check", k)
                    print()
                    k += 1
                    solution.showPlace()
                    solution.travelToBoat("Origin", "Boat", )
                    solution.showPlace()
                    solution.travelToDestination()
                    solution.showPlace()


                    if solution.checkEmpty("Boat") == False and solution.checkEmpty("Origin") == False: # if boat is not empty and origin is not empty, bring back a passenger to origin
                        solution.travelToOrigin()
                        solution.showPlace()


                    elif solution.checkEmpty("Boat") and solution.checkEmpty("Origin") == False: # if boat is empty and origin is not empty, bring back a passenger to the boat into origin
                        solution.travelBackToBoat()
                        solution.showPlace()
                        solution.travelToOrigin()
                        solution.showPlace()


                    if solution.checkEmpty("Boat") == False and solution.checkEmpty("Origin"): # if there is only one passenger left in the boat and no one is at origin, travel the passenger to destination
                        solution.travelToDestination()
                        solution.showPlace()
                    print()
                    print()
                    print()
```

```python
outer = river_crossing()
outer.execute()
```
