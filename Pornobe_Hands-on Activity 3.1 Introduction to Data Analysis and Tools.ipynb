```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

### Mean

```python
#Mean without import from statistics module
sum = 0
for i in salaries:
    sum += i
meanAns = sum/len(salaries)
print(f"Mean: {meanAns}")
```

```python
#Mean with importing from statistics module
from statistics import mean

print(f"Mean: { mean(salaries)}")
```

### Median

```python
#Median without import from statistics module
#sort the the data before getting the median (just in case the data is unsorted)

sortedList = sorted(salaries)
length = len(salaries)

if (length % 2) == 1:               #if the length of the list is odd get the (n/2 + 1)
    
    medianLoc = int((length + 1) / 2)          # to ge the nth place of the median //note: i added int() because the list cannot have a float index
    print(f"The median is: {sortedList[medianLoc]}")

else:                                     # if even get the average of the n/2 term and the n/2 + 1 term
    x = int(length / 2)
    median = (sortedList[x-1] + sortedList[x]) / 2
    print(median)
```

```python
#Median with importing from statistics module
from statistics import median
print(f"Median: ", median(salaries))
```

### Mode

```python
#Mode without importing from statistics module
from collections import Counter

Counter(salaries).most_common(1) #outputs the most common in the Counter dictionary subclass (key, value)
                                 #, the value is the amount of occurence in the data
```

```python
#Mode with importing from statistics module
from statistics import mode
print(f"Mode: {mode(salaries)}")
```

### Sample Variance

```python
#Sample Variance without importing from statistics module
x = 0
for i in salaries:               
    x += (i - meanAns)**2         #get the summation of (xi - mean)^2 //I used the mean from the earlier output to shorten the code
    
sVariance = x / (len(salaries) - 1)     #then divide by n - 1 where n is the size of the data
print(f"Sample variance: {sVariance}")
```

```python
#Sample Variance with importing from statistics module
from statistics import variance

print(f"Sample variance: {variance(salaries, meanAns)}")        # added the mean to prevent recalculation of the mean
```

### Sample Standard Deviation

```python
#Sample Standard Deviation without importing from statistics module
import math
x = 0
for i in salaries:               
    x += (i - meanAns)**2         #get the summation of (xi - mean)^2 //I used the mean from the earlier output
    
sVariance = x / (len(salaries) - 1)     #then divide by n - 1 where n is the size of the data
standardDeviation = math.sqrt(sVariance)     #i used the math module for a short code
print(f"Sample Standard Deviation: {standardDeviation}")

```

```python
#Sample Standard Deviation with importing from statistics module
from statistics import stdev

print(f"Sample Standard Deviation: {stdev(salaries)}")        
```

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:
 1. Range

### Range

```python
# Range
rnge = max(salaries) - min(salaries) # get the difference between the max value and minimum value of the dataset
print(f"Range: {rnge}")
```

### Coefficient of variation Interquartile range


```python
# Coefficient of variation Interquartile range
import statistics as st

quantilesList = st.quantiles(salaries, n = 4) #get the list of Q1 - Q3
IQR = quantilesList[2] - quantilesList[0]     # get the interquartile range IQR = Q3 - Q1
median = st.median(salaries)

cvIQR = (IQR / median) * 100           #use the formula of CV but instead of stdev and mean, use iqr and median
print(f"CV IQR: {cvIQR}%")

```

*High Coefficient of Variation based on Interquartile range. Therefore, the salaries are greatly spread out or varies from each
other compared to the median value.*

### Quartile coefficient of dispersion

```python
# Quartile coefficient of dispersion
import statistics as st

quantilesList = st.quantiles(salaries, n = 4)                                          #get the list of Q1 - Q3 again, then
QCD = (quantilesList[2] - quantilesList[0])/(quantilesList[2] + quantilesList[0])      # find the quotient of IQR and Q3 + Q1
print(f"QCD: {QCD}%")

```

*Low Quartile Coefficient of Dispersion. Therefore, the middle values are not greatly varied. The employees are earning similar amounts in the middle part of the data.*
