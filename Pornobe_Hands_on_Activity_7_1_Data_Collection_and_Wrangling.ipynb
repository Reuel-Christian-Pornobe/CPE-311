# Exercise 1

```python
# Read each file in
import pandas as pd

nflx, goog, fb, amzn, aapl = pd.read_csv('nflx.csv'), pd.read_csv('goog.csv'), pd.read_csv('fb.csv'), pd.read_csv('amzn.csv'), pd.read_csv('aapl.csv')
```

```python
# 2. add a column named ticker

def addTicker(dataFrame, tickerName):
    dataFrame['ticker'] = tickerName

addTicker(aapl, 'AAPL')
addTicker(amzn, 'AMZN')
addTicker(fb, 'FB')
addTicker(goog, 'GOOG')
addTicker(nflx, 'NFLX')
```

```python
# append them together in a single data frame
df = pd.concat([aapl, amzn, fb, goog, nflx])
df
```

```python
# save the dataframe into a csv

df.to_csv('faang.csv')
```

# Exercise 2

```python
# use type conversion to change date column to a date_time and volume to integer. Then, sort by date and ticker

df.date = df.date.apply(pd.to_datetime)          # change date column to datetime
df.volume = df.volume.apply(pd.to_numeric)       # change volume column to integer


df.sort_values(['date', 'ticker'])               #sort the dataframe by date and ticker
```

```python
# find the seven rows with the highest value for volume
df.nlargest(7, 'volume')
```

```python
# melt the dataframe to merge open, high, low,close, and volume
dfMelted = df.melt(id_vars = ['date', 'ticker'], value_vars = ['open', 'high', 'low', 'close', 'volume'],
                   var_name = 'stock category', value_name = 'price' )            # i am not sure if the open, high, low, close, etc. values means price in stocks (i based it from google)
dfMelted
```

# Exercise 3

```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_hospitals_in_the_Philippines'              #the url in which the data is stored
page = requests.get(url)                                                                # get the data from the url

soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
table = soup.find_all('table')[0]    # get the specific table: I was able to know which table I need by using the inspect in the website
```

```python
wrldTitles = table.find_all('th')           #extract the table titles from the table
wrldTitles
```

```python
wrldTableTitles = [title.text.strip() for title in wrldTitles]     # make it a title.text while also stripping the </th>
wrldTableTitles
```

```python
import pandas as pd
df = pd.DataFrame(columns = wrldTableTitles)              # put the extracted titles into a dataframe
df
```

```python
columnData = table.find_all('tr')              #extracting the data of the table
```

```python
for row in columnData[1:]:         #used the [1:] because the first row has an empty list which will mismatch with the columns
    rowData = row.find_all('td')                               #find all individual data
    individualRowData = [data.text.strip() for data in rowData]  #get all the row data by finding all with the 'td' then strip the /n
    df.loc[len(df)] = individualRowData
```

```python
df
```

```python
# save the df to a csv file
df.to_csv('hostpital.csv')
```
