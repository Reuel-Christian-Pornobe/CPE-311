```python
import re

nmap = df2.query('Attack_type.str.contains("^NMAP")')             # all nmap scans
nmap.shape
```

```python
other = attack.query('~Attack_type.str.contains("^NMAP")')        # other attack types
other.shape
```

```python
fin = attack.query('Attack_type == "NMAP_FIN_SCAN"')        # FIN SCAN attack types
fin.shape
```

### NMAP amount of attack

```python
nmap.Attack_type.value_counts()
```

FIN scan is the least frequent nmap scan attack

### NMAP IAT.avg

```python
nmap.groupby('Attack_type')['fwd_iat.avg'].mean().sort_values()
```

The nmap OS detection pattern has the quickest iat while UDP scan pattern is the slowest time between intervals.

```python
nmap.groupby('Attack_type')['bwd_iat.avg'].mean().sort_values()
```

The nmap OS detection pattern still has the quickest iat while the NMAP FIN scan pattern is now the slowest time between intervals.

### NMAP payload bytes per second

```python
nmap.groupby('Attack_type')['payload_bytes_per_second'].mean().sort_values()
```

UDP SCAN has the highest payload bytes per second while OS detection has no payload maybe due to the simplicity of the scan, which is just detecting the Operating System of the device

### NMAP triggered flags

```python
nmap.Attack_type.value_counts()
```

Fin scan is almost undetected by the flags except the finish flag

# Visualization

## Distribution of all Attack type

```python
attack_counts = attack['Attack_type'].value_counts()

plt.figure(figsize=(10, 6))                         # size of the canvas(??) of the plot
plt.bar(attack_counts.index, attack_counts.values)

# layout fixes
plt.xticks(rotation=45, ha='right')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.title('Count of Each Attack Type (excluding the normal)')
plt.tight_layout()
plt.show()
```

## Check the distribution of NMAP types

```python
attack_counts_nmap = nmap['Attack_type'].value_counts()

plt.figure(figsize=(10, 6))                         # size of the canvas(??) of the plot
plt.bar(attack_counts_nmap.index, attack_counts_nmap.values)

# layout fixes
plt.xticks(rotation=45, ha='right')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.title('Count of Each Nmap Attack Type')
plt.tight_layout()
plt.show()
```

NMAP_FIN_SCAN is the least used NMAP scan.

## Flags triggered by NMAP SCANS

```python
flags = fin.groupby('Attack_type')[['flow_FIN_flag_count','flow_SYN_flag_count', 'flow_RST_flag_count', 'fwd_PSH_flag_count',
                             'flow_ACK_flag_count', 'fwd_URG_flag_count']].sum()

plt.figure(figsize=(10, 6))                         # size of the canvas(??) of the plot
flags.plot(kind = 'bar')

# layout fixes
plt.xticks(rotation = 0)
plt.title('Count of Each Flags triggered by NMAP FIN SCAN')
plt.tight_layout()
plt.show()
```

NMAP FIN_SCAN usually triggers the FIN_flag_count

```python
flags = nmap.query('Attack_type == "NMAP_XMAS_TREE_SCAN"').groupby('Attack_type')[['flow_FIN_flag_count','flow_SYN_flag_count', 'flow_RST_flag_count', 'fwd_PSH_flag_count',
                             'flow_ACK_flag_count', 'fwd_URG_flag_count']].sum()

plt.figure(figsize=(10, 6))                         # size of the canvas(??) of the plot
flags.plot(kind='bar')

# layout fixes
plt.xticks(rotation = 0)
plt.title('Count of Each Flags triggered by NMAP XMAS TREE SCAN')
plt.tight_layout()
plt.show()
```

```python
flags1 = nmap.query('Attack_type == "NMAP_XMAS_TREE_SCAN"').groupby('Attack_type')[['flow_FIN_flag_count','flow_SYN_flag_count', 'flow_RST_flag_count', 'fwd_PSH_flag_count',
                             'flow_ACK_flag_count', 'fwd_URG_flag_count']].sum()


aq = flags1.plot(kind='bar', rot = 0, figsize = (20,10))
aq.legend(fontsize = 18)
for container in aq.containers:
    for bar in container:
        height = bar.get_height()
        aq.text(bar.get_x() + bar.get_width() / 2, height + 50, f'{height:.0f}',
                ha='center', va='bottom', fontsize=20)

plt.show()
```
