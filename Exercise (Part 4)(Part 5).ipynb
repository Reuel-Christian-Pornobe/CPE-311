1.   Using the meteorite data from the Meteorite_Landings.csv file, create a pivot table that shows both the number of meteorites and the 
95th percentile of meteorite mass for those that were found versus observed falling per year from 2005 through 2009 (inclusive). 
Hint: Be sure to convert the year column to a number as we did in the previous exercise


```python
import pandas as pd
df = pd.read_csv('Meteorite_Landings.csv')
df
```

```python
# converting year to a number
df['year'] = df.year.str.slice(6, 10).apply(pd.to_numeric)
```

```python
df
```

```python
#create a pivot table

df['sortedYear'] = df.year[(df.year >= 2005) & (df.year <= 2009)]        #add a column in df with the sorted year

pivot = df.pivot_table(
    index='sortedYear', columns='fall', sort=True, values = 'mass (g)', 
    aggfunc = {'mass (g)': ['count', lambda i: i.quantile(0.95)]}         #count the meteorites then use lambda to output the 95th percentile    
)
pivot = pivot.fillna(0)          #fill all the NaN values with 0
pivot
```

1. Using the taxi trip data in the 2019_Yellow_Taxi_Trip_Data.csv file, resample the data to an hourly frequency based on the dropoff time. Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount, then find the 5 hours with the most tips.

```python
taxis = pd.read_csv('2019_Yellow_Taxi_Trip_Data.csv')
taxis
```

```python
#convert the drop off time to date time

taxis['tpep_dropoff_datetime'] = taxis.tpep_dropoff_datetime.apply(pd.to_datetime)

```

```python
#convert the pickup time to date time
taxis['tpep_pickup_datetime'] = taxis.tpep_pickup_datetime.apply(pd.to_datetime)
```

```python
# set the index to drop off for the hourly frequency
taxis = taxis.set_index('tpep_dropoff_datetime').sort_index()
```

```python
# for sorting the hourly frequency
taxis = taxis.sort_values('tpep_dropoff_datetime', ascending = True)
```

```python
#  Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount
taxis[['trip_distance', 'fare_amount', 'tip_amount']].resample('60min').sum()
```

```python
# did not finish the calculate 5 hours
taxis['5hours'] = taxis['tpep_pickup_datetime'] - taxis['tpep_dropoff_datetime']
```
